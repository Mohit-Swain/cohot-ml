```python
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def makeGrayScale(text_dir,gray_dir = 'gray-data'):
    text_dir = 'data'
    if not os.path.exists(text_dir):
        raise Exception('no such directory Exists')
    if os.path.exists(gray_dir):
        raise Exception(gray_dir+' directory already present')
    os.mkdir(gray_dir)
    img_path_list = list(filter(lambda x : os.path.splitext(x)[-1] == '.jpg',os.listdir(text_dir)))
    print('Total number of images',len(img_path_list))
    for image_name in img_path_list:
        imageBGR = cv2.imread(os.path.join(text_dir,image_name))
        image = cv2.cvtColor(imageBGR,cv2.COLOR_BGR2RGB)
        print(image_name)
        f,axarr = plt.subplots(1,2)
        axarr[0].imshow(image)
        axarr[0].axis("off")
        gray_image = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
        axarr[1].imshow(gray_image,cmap='gray', vmin=0, vmax=255)
        axarr[1].axis("off")
        plt.show()
        cv2.imwrite(os.path.join('grayData',image_name),gray_image)
```

```python
makeGrayScale('data')
```
